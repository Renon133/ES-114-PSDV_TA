### 1. List Operations and Basic Statistics
1. Create a list of the first 20 natural numbers.
2. Compute the sum of all even numbers in the list.
3. Compute the product of all odd numbers in the list.
4. Calculate the mean, median, and standard deviation of the list.

```python
import numpy as np
from math import prod

# 1. Create a list of the first 20 natural numbers.
natural_numbers = list(range(1, 21))

# 2. Compute the sum of all even numbers in the list.
sum_even = sum(num for num in natural_numbers if num % 2 == 0)

# 3. Compute the product of all odd numbers in the list.
product_odd = prod(num for num in natural_numbers if num % 2 != 0)

# 4. Calculate the mean, median, and standard deviation of the list.
mean = np.mean(natural_numbers)
median = np.median(natural_numbers)
std_dev = np.std(natural_numbers)

print(f"Sum of even numbers: {sum_even}")
print(f"Product of odd numbers: {product_odd}")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
```

2. Provide examples of each.

- Mutable data types are those that can be changed after their creation.
Examples include lists, dictionaries, and sets.
Immutable data types are those that cannot be changed after their creation.
Examples include integers, floats, strings, and tuples.

### 3. Fibonacci Sequence
1. Write a Python function that takes a number `n` and returns a list of the first `n` Fibonacci numbers.
2. Plot the Fibonacci sequence using a line graph.

```python
import matplotlib.pyplot as plt

def fibonacci(n):
    fib_sequence = [0, 1]
    while len(fib_sequence) < n:
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence[:n]

# Generate the first 20 Fibonacci numbers
n = 20
fib_sequence = fibonacci(n)

# Plot the Fibonacci sequence
plt.plot(fib_sequence, marker='o')
plt.title(f'First {n} Fibonacci Numbers')
plt.xlabel('Index')
plt.ylabel('Fibonacci Number')
plt.grid(True)
plt.show()
```

 ### 6. Palindrome Check
1. Write a program to check if a string is a palindrome (ignoring spaces and case).
    - Input: "A man a plan a canal Panama"
    - Output: True

```python
def is_palindrome(s):
    # Remove spaces and convert to lowercase
    cleaned_str = ''.join(s.split()).lower()
    # Check if the cleaned string is equal to its reverse
    return cleaned_str == cleaned_str[::-1]

# Test the function
input_str = "A man a plan a canal Panama"
print(is_palindrome(input_str))  # Output: True
```

### 9. Merging Dictionaries
1. Write Python code to merge two dictionaries. If a key is present in both, sum their values.
    - Example:
        - dict1 = {'a': 10, 'b': 20}
        - dict2 = {'b': 5, 'c': 15}
    - Output: {'a': 10, 'b': 25, 'c': 15}

```python
def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()  # Start with dict1's keys and values
    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value  # If key is in both, sum the values
        else:
            merged_dict[key] = value  # If key is only in dict2, add it to merged_dict
    return merged_dict

# Example dictionaries
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}

# Merging the dictionaries
merged_dict = merge_dictionaries(dict1, dict2)
print(merged_dict)  # Output: {'a': 10, 'b': 25, 'c': 15}
```


 ### 10. Character Frequency
1. Write a function that accepts a string and:
    - Converts all characters to lowercase.
    - Counts the frequency of each character.
    - Returns a dictionary with characters as keys and their counts as values.
2. Visualize the character frequencies using a bar chart.

```python
def character_frequency(s):
    # Convert all characters to lowercase
    s = s.lower()
    # Count the frequency of each character
    freq_dict = {}
    for char in s:
        if char in freq_dict:
            freq_dict[char] += 1
        else:
            freq_dict[char] = 1
    return freq_dict

# Example string
example_string = "A man a plan a canal Panama"

# Get the character frequency dictionary
char_freq = character_frequency(example_string)
if ' ' in char_freq:
    char_freq['space'] = char_freq[' ']
    del char_freq[' ']
    
print(char_freq)

# Visualize the character frequencies using a bar chart
plt.bar(char_freq.keys(), char_freq.values(), color='green')
plt.xlabel('Characters')
plt.ylabel('Frequency')
plt.title('Character Frequency in the String')
plt.show()
```

 ### 11. Remove Duplicates
1. Write a program to remove duplicate elements from a list while maintaining the order of elements.
2. Calculate the mean and median of the unique elements.

```python
# Given list with duplicates
duplicate_list = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7, 8, 8, 9, 10, 10]

# 1. Remove duplicate elements while maintaining the order
unique_list = []
seen = set()
for item in duplicate_list:
    if item not in seen:
        unique_list.append(item)
        seen.add(item)

# 2. Calculate the mean and median of the unique elements
mean_unique = np.mean(unique_list)
median_unique = np.median(unique_list)

print(f"Unique list: {unique_list}")
print(f"Mean of unique elements: {mean_unique}")
print(f"Median of unique elements: {median_unique}")
```

 ### 12. String Operations
1. Given a list of strings, write a Python function to:
    - Find the longest string.
    - Count how many strings start with a vowel.
    - Visualize the distribution of string lengths using a histogram.

```python
def string_operations(strings):
    # Find the longest string
    longest_string = max(strings, key=len)
    
    # Count how many strings start with a vowel
    vowels = 'aeiouAEIOU'
    count_vowel_start = sum(1 for s in strings if s and s[0] in vowels)
    
    # Visualize the distribution of string lengths using a histogram
    string_lengths = [len(s) for s in strings]
    plt.hist(string_lengths, bins=20, edgecolor='black')
    plt.title('Distribution of String Lengths')
    plt.xlabel('Length of String')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
    
    return longest_string, count_vowel_start

# Example list of strings
strings = ["apple", "banana", "cherry", "date", "elderberry", "fig", "grape"]

# Perform string operations
longest_string, count_vowel_start = string_operations(strings)
print(f"Longest string: {longest_string}")
print(f"Number of strings starting with a vowel: {count_vowel_start}")
```


### 13. Sorting Dictionaries
1. Write a function to sort a list of dictionaries by a key.
    - Example:
        - data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
        - Sort by age → [{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}]

```python
def sort_dicts_by_key(data, key):
    return sorted(data, key=lambda x: x[key])

# Example data
data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]

# Sort by age
sorted_data = sort_dicts_by_key(data, 'age')
print(sorted_data)  # Output: [{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}]
```

 ### 14. Word Filtering and Sorting
1. Write a program to:
    - Split a string into a list of words.
    - Remove all words that are shorter than 4 characters.
    - Sort the remaining words alphabetically.
    - Visualize the word lengths using a bar chart.

```python
def filter_and_sort_words(input_string):
    # Split the string into a list of words
    words = input_string.split()
    
    # Remove all words that are shorter than 4 characters
    filtered_words = [word for word in words if len(word) >= 4]
    
    # Sort the remaining words alphabetically
    sorted_words = sorted(filtered_words)
    
    # Visualize the word lengths using a bar chart
    word_lengths = [len(word) for word in sorted_words]
    plt.bar(sorted_words, word_lengths, color='purple')
    plt.xlabel('Words')
    plt.ylabel('Length')
    plt.title('Word Lengths')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
    
    return sorted_words

# Example input string
input_string = "This is an example string with several words of different lengths"

# Perform word filtering and sorting
sorted_words = filter_and_sort_words(input_string)
print(f"Sorted words: {sorted_words}")
```

### 15. Anagram Finder
1. Write a Python function to find all anagrams of a word in a given list.
    - Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]
    - Output: ["enlist", "inlets"]

```python
def find_anagrams(word, words):
    sorted_word = sorted(word)
    return [w for w in words if sorted(w) == sorted_word]

# Example input
word = "listen"
words = ["enlist", "google", "inlets", "banana"]

# Find anagrams
anagrams = find_anagrams(word, words)
print(anagrams)  # Output: ["enlist", "inlets"]
```

### 16. Set Operations
1. Given two lists, write Python code to find their intersection, union, and symmetric difference.
2. Visualize the results using Venn diagrams.

```python
from matplotlib_venn import venn2

# Given lists
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]

# Convert lists to sets
set1 = set(list1)
set2 = set(list2)

# Find intersection, union, and symmetric difference
intersection = set1 & set2
union = set1 | set2
symmetric_difference = set1 ^ set2

print(f"Intersection: {intersection}")
print(f"Union: {union}")
print(f"Symmetric Difference: {symmetric_difference}")

# Visualize the results using Venn diagrams
venn2([set1, set2], ('Set 1', 'Set 2'))
plt.title('Venn Diagram of Set Operations')
plt.show()
```

### 18. Word Count and Frequency
1. Write a program to:
    - Read a file and count the number of words in it.
    - Find the most frequent word.
    - Visualize the word frequencies using a bar chart.

```python
from collections import Counter

# Read the file and count the number of words
with open('numbers.txt', 'r') as file:
    text = file.read()
    words = text.split()
    word_count = len(words)

# Find the most frequent word
word_freq = Counter(words)
most_frequent_word = word_freq.most_common(1)[0]

print(f"Number of words in the file: {word_count}")
print(f"Most frequent word: {most_frequent_word[0]} (Frequency: {most_frequent_word[1]})")

# Visualize the word frequencies using a bar chart
plt.bar(word_freq.keys(), word_freq.values(), color='blue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequencies')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### 22. Normalization
1. Write a Python function to normalize a list of numbers to a range of 0 to 1.
2. Visualize the normalized data using a line plot.

```python
def normalize(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    return [(num - min_val) / (max_val - min_val) for num in numbers]

# Normalize the numbers from the DataFrame
numbers = [27, 10, 45, 45, 19, 34, 40, 40, 5]
print('Numbers:' , numbers)
normalized_numbers = normalize(numbers)
print('Normalized Numbers:' , normalized_numbers)

# Visualize the normalized data using a line plot
plt.plot(normalized_numbers, marker='o')
plt.title('Normalized Data')
plt.xlabel('Index')
plt.ylabel('Normalized Value')
plt.grid(True)
plt.show()
```

### 24. Grade Calculator
1. Write a Python program that takes a numerical grade (0-100) as input and prints the corresponding letter grade (A, B, C, D, F).

```python
# Function to convert numerical grade to letter grade
def grade_calculator(numerical_grade):
    if 90 <= numerical_grade <= 100:
        return 'A'
    elif 80 <= numerical_grade < 90:
        return 'B'
    elif 70 <= numerical_grade < 80:
        return 'C'
    elif 60 <= numerical_grade < 70:
        return 'D'
    else:
        return 'F'

# Take a numerical grade as input from the user
numerical_grade = int(input("Enter a numerical grade (0-100): "))

# Get the corresponding letter grade
letter_grade = grade_calculator(numerical_grade)

# Print the letter grade
print(f"The letter grade for {numerical_grade} is {letter_grade}.")
```

 ### 25. Leap Year Checker
1. Write a Python program that takes a year as input and checks if it is a leap year.


```python
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# Take a year as input from the user
year = int(input("Enter a year: "))

# Check if the year is a leap year
if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```
